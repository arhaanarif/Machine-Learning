```python
from faker import Faker
import pandas as pd

# Initialize Faker
fake = Faker()

# Function to generate fake data
def generate_fake_data(num_rows):
    data = []
    for _ in range(num_rows):
        age = fake.random_int(min=22, max=65)  # Age between 22 and 65
        salary = fake.random_int(min=30000, max=150000)  # Salary between 30,000 and 150,000
        experience = fake.random_int(min=0, max=age-21)  # Experience between 0 and (Age - 21)
        data.append({
            'Age': age,
            'Salary': salary,
            'Experience': experience
        })
    return data

# Generate fake data for 100 rows
fake_data = generate_fake_data(100)


```

```python
# Convert to DataFrame for easier handling
df = pd.DataFrame(fake_data)

# Display the first few rows
df.head(10)
```

```python
df.to_csv('fake_data.csv',index=False)

#! to save the data in csv format
```
