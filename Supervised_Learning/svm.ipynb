#### SVM: Supportg Vector Machine

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
dataset=pd.read_csv("C:/SDE/ML/LogRe_FakeData.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa",y="score",data=dataset,hue="placed")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placed"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC(kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
```
