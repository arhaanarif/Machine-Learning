#### How to handle duplicate data

```python
import pandas as pd
```

```python
data={
    "name":["a","b","c","d","a","c"],
    "eng":[8,7,5,8,8,4],
    "hindi":[2,3,4,5,2,6]
}

df=pd.DataFrame(data)
df
```

```python
df.drop_duplicates(inplace=True)
df

#! df.duplicates()  !!to check for duplicates
```
